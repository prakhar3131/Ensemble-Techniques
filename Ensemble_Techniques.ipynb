  Ensemble Learning in ML refers to use of multiple models for making the prediction so as to make the model more stable and accurate.It helps i making tge model more robust.

2.What is the difference between Bagging and Boosting?
  BAGGING- It is the process of training models parallely in which models are independent of each other and it takes diff samples with replacement for training for each model.It is also called Bootstrap Aggregation.
  
  BOOSTING- It is the process of training models sequentially one after the other that are dependent on each other and learn from the mistakes of previous models.

3.What is bootstrap sampling and what role does it play in Bagging methods like Random Forest?
  Bootstrap sampling or bagging the model is trained by taking diff samples with replacement.
  It helps in random forest algorithm by randmoly sampling the rows and features then splitting the data into small chunks thus helps in reducing the overfitting.

4.What are Out-of-Bag OOB samples and how is OOB score used to evaluate ensemble models?
  Out-of-Bag sample is the part of training data not used for training for each individual model.The data whcih is no used for training is used to validate the each secific individual DT. Overall oob score for entire model is taken by taking the average of oob scores of each individual decision trees.
